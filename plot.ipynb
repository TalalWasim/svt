```python
import os
import matplotlib.pyplot as plt
import ast

%matplotlib inline
```

```python
results_path = './results'
```

```python
plt.figure(figsize=(14,7))

for result in os.listdir(results_path):
    if result.startswith('.'):
        continue
    file_path = os.path.join(results_path, result, 'log.txt')
    
    with open(file_path, 'r') as f:
        data = f.readlines()
    
    plot_list = []
        
    for line in data:
        line_dict = ast.literal_eval(line.strip())
        plot_list.append(line_dict['train_loss'])
    
    plt.plot(plot_list, label = result)
plt.xlabel('Epochs', fontsize=20)
plt.ylabel('Loss', fontsize=20)
plt.legend()
plt.show()
```
